# 1. BGM音频分析

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 设置画图的中文显示
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(){return false} 
```

```python
# 导入数据文件
io = "../../data/BGMAudio.xlsx"
```

```python
# 获取BGM音频数据

# 实验组
audio = pd.read_excel(io,sheet_name = "Sheet1",index_col= 'BGM')

# 对照组1
controlGroup1 = pd.read_excel(io,sheet_name = "对照组1",index_col= 'BGM')

# 对照组2
controlGroup2 = pd.read_excel(io,sheet_name = "对照组2",index_col= 'BGM')
```

```python
audio
```

```python
controlGroup1
```

```python
controlGroup2
```

```python
# 展示实验组的数据

audio.plot(kind='bar',figsize=(20,8),grid=True,rot=0)
```

```python
# 展示对照组1的数据

controlGroup1.plot(kind='bar',figsize=(20,8),grid=True,rot=0)
```

```python
# 展示对照组2的数据

controlGroup2.plot(kind='bar',figsize=(20,8),grid=True,rot=0)
```

```python
# 实验组的数据展示
audio.T.plot.bar(figsize=(20,8),grid=True,rot=0)
```

```python
# 对照组1的数据展示
controlGroup1.T.plot.bar(figsize=(20,8),grid=True,rot=0)
```

```python
# 对照组2的数据展示
controlGroup2.T.plot.bar(figsize=(20,8),grid=True,rot=0)
```

### 将实验组和对照组的各项数据取平均值进行对比

```python
compare = pd.DataFrame({"实验组":audio.mean(),"对照组1":controlGroup1.mean(),"对照组2":controlGroup2.mean()})
```

```python
compare
```

```python
# 实验组和对照组的数据对比展示
compare.plot.bar(figsize=(20,8),grid=True,rot=0)
```
